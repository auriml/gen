```python
import h5py
```

```python
#Importing TCGA RNA expression data downloaded from  https://amp.pharm.mssm.edu/archs4/download.html
filename = "tcga_matrix.h5"

h5 = h5py.File(filename,'r')
for k in h5.keys():
    print(k)


meta = h5['meta']  
data = h5['data']  
info = h5['info']  



#Showing which type of data it contains    
for s in meta.keys():
    print(s)
    print(meta[s].shape)
    print(meta[s].value)
    
for s in data.keys():
    print(s)
    print(data[s].shape)
    print(data[s].value)    
    
for s in info.keys():
    print(s)
    print(info[s].shape)
    print(info[s].value)    

#h5.close()
```
